# **Clase 02 - Árboles de decisión**

En este ejercicio, aprenderemos a entrenar árboles de decisión usando `sklearn`.


Primero, vamos a importar nuestro conjunto de datos


```python
%matplotlib inline
import numpy as np
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target

print("Class labels:", np.unique(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y
)
```

```python
print("Labels counts in y:", np.bincount(y))
print("Labels counts in y_train:", np.bincount(y_train))
print("Labels counts in y_test:", np.bincount(y_test))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from mlxtend.plotting import plot_decision_regions
```

```python
tree = DecisionTreeClassifier(criterion="gini", random_state=1)
tree.fit(X_train, y_train)


plot_decision_regions(X_train, y_train, tree)

plt.xlabel("petal length [cm]")
plt.ylabel("petal width [cm]")
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(10, 7))

plot_tree(
    tree,
    filled=True,
    rounded=True,
    class_names=["Setosa", "Versicolor", "Virginica"],
    feature_names=["petal length", "petal width"],
)

plt.show()
```
